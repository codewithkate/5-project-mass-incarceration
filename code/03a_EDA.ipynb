```python
#reading in file 
booker=pd.read_csv('datasets/clean_booker.csv')
booker.head()
```

```python
def fill_values(cols:list,fill,value = 0, nans = False):
    """
    Takes a list of columns and maps value to fill in missing values. 
    """
    if nans == True:
        for col in cols:
            booker[col]= booker[col].apply(lambda x: value if np.isnan(x) else x)
    else:
        for col in cols:
            booker[col]= booker[col].map({fill:value})

        
    
```
